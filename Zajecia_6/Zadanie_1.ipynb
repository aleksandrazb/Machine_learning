```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import numpy as np

data = pd.read_csv('data6.tsv', sep='\t', header=None)

plt.figure(figsize=(10,6))
plt.scatter(data[0], data[1])
plt.show()
```

```python
X = pd.DataFrame.from_dict(data[0])
y = pd.DataFrame.from_dict(data[1])
X_seq = np.linspace(X.min(),X.max(),300).reshape(-1,1)

def polynomial(degree):
    polyreg = make_pipeline(PolynomialFeatures(degree),LinearRegression())
    polyreg.fit(X,y)
    y_predicted = polyreg.predict(X_seq)
    return y_predicted

y_predicted1 = polynomial(1)
y_predicted2 = polynomial(2)
y_predicted5 = polynomial(5)

plt.figure(figsize=(10,6))
plt.scatter(data[0], data[1])
plt.plot(X_seq, y_predicted1, c='red')
plt.plot(X_seq, y_predicted2, c='green')
plt.plot(X_seq, y_predicted5, c='blue')
plt.show()
```
